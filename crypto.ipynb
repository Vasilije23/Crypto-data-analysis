```python
btc = pd.read_csv("BTC.csv")
btc = btc.drop(labels=range(0, 2732), axis=0)
btc.head()
```

```python
btc['Date'] = pd.to_datetime(btc['Date'])
btc = btc.set_index('Date')
btc.head()
```

```python
btc.describe()
```

```python
btc['Closing Price (USD)'].plot(figsize=(10,6), label='Bitcoin Price')
plt.legend()
plt.ylabel('Price in $')
plt.show()
```

```python
btc['24h Low (USD)'].plot(label='24h Low (USD)', figsize=(12,8))
btc['24h High (USD)'].plot(label='24 High (USD)', figsize=(12,8))
plt.legend()
plt.title('24 Low vc High price (USD)')
plt.ylabel('Price in $')
plt.show()
```

```python
btc['24h Open (USD)'].plot(label='24h Open (USD)', figsize=(12,8))
btc['Closing Price (USD)'].plot(label='Closing Price (USD)', figsize=(12,8))
plt.legend()
plt.ylabel('Price in $')
plt.show()
```

```python
crypto = pd.DataFrame({'BTC': btc['Closing Price (USD)'], 'DOT': dot['Closing Price (USD)'], 'EOS': eos['Closing Price (USD)'], 'XTZ': xtz['Closing Price (USD)']})

crypto.head()
```

```python
crypto.tail()
```

```python
crypto.describe()
```

```python
crypto_return = crypto.apply(lambda x: (x / x[0] - 1) * 100)
crypto_return.head()
```

```python
crypto_return.tail()
```

```python
crypto_return.plot( figsize=(14, 10)).axhline(y = 0, color = "black", lw = 1.5)
plt.title('Returns')
plt.ylabel('Return in %')
```
